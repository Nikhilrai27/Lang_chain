```python
import langchain
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool

def get_weather(location:str)->str:
    """GET WEATHER AT LOCATION"""
    return f"the weather at {location} is sunny"
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)
```

```python
#binding the model
llm_with_tools = model.bind_tools([get_weather])
```

```python
response=llm_with_tools.invoke("whats weather in Lucknow")
print(response)
```

```python
response.tool_calls
```

```python
from langchain.agents import create_agent
```

```python
agent=create_agent(
    model=model,
    tools=[get_weather],
    system_prompt="helpful assitant"
)
```

```python
result=agent.invoke(
        {"messages": [{"role": "user", "content": "What's the weather in San Francisco?"}]}
)
```
